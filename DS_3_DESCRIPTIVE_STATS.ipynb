```python
def calculate_mean_using_sum(data):
    """
    Calculate the mean of a numerical feature in a dataset using sum.

    Parameters:
    - data: Series, the numerical feature for which to calculate the mean

    Returns:
    - float: the mean value
    """
    sum_values = sum(data)
    count_values = len(data)
    return sum_values / count_values

def calculate_median_using_sort(data):
    """
    Calculate the median of a numerical feature in a dataset using sorting.

    Parameters:
    - data: Series, the numerical feature for which to calculate the median

    Returns:
    - float: the median value
    """
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 0:
        return (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
    else:
        return sorted_data[n // 2]

def calculate_minimum_using_min(data):
    """
    Calculate the minimum value of a numerical feature in a dataset using the min() function.

    Parameters:
    - data: Series, the numerical feature for which to calculate the minimum value

    Returns:
    - float: the minimum value
    """
    return min(data)

def calculate_maximum_using_max(data):
    """
    Calculate the maximum value of a numerical feature in a dataset using the max() function.

    Parameters:
    - data: Series, the numerical feature for which to calculate the maximum value

    Returns:
    - float: the maximum value
    """
    return max(data)

def calculate_standard_deviation_using_sum(data):
    """
    Calculate the standard deviation of a numerical feature in a dataset using the sum method.

    Parameters:
    - data: Series, the numerical feature for which to calculate the standard deviation

    Returns:
    - float: the standard deviation value
    """
    mean = calculate_mean_using_sum(data)
    squared_diff = sum((x - mean) ** 2 for x in data)
    return (squared_diff / len(data)) ** 0.5

# Example usage:
tip_mean_using_sum = calculate_mean_using_sum(tips['tip'])
tip_median_using_sort = calculate_median_using_sort(tips['tip'])
tip_min_using_min = calculate_minimum_using_min(tips['tip'])
tip_max_using_max = calculate_maximum_using_max(tips['tip'])
tip_std_using_sum = calculate_standard_deviation_using_sum(tips['tip'])

print("Mean tip amount using sum method:", tip_mean_using_sum)
print("Median tip amount using sort method:", tip_median_using_sort)
print("Minimum tip amount using min method:", tip_min_using_min)
print("Maximum tip amount using max method:", tip_max_using_max)
print("Standard deviation of tip amount using sum method:", tip_std_using_sum)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Calculate summary statistics
tip_mean = tips['tip'].mean()
tip_median = tips['tip'].median()
tip_min = tips['tip'].min()
tip_max = tips['tip'].max()
tip_std = tips['tip'].std()

# Create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=tips['tip'])
plt.title('Summary Statistics of Tip Amount')
plt.ylabel('Tip Amount')
plt.xlabel('Summary Statistics\nMean: {:.2f}, Median: {:.2f}, Min: {:.2f}, Max: {:.2f}, Std: {:.2f}'.format(tip_mean, tip_median, tip_min, tip_max, tip_std))
plt.show()
```
